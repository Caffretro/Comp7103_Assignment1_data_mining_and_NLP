```python
import json
import pandas as pd

path = './data_amazon_review/Musical_instruments_reviews.csv'

user_data = pd.read_csv(path)
user_data.head()
```

### Based on user ratings in the 'overall' column, label reviews with Positive, Neutral, and Negative.

```python
user_data['review'] = user_data['reviewText'] + user_data['summary']
user_data.drop(columns=['reviewText', 'summary'], inplace=True)
user_data.head()
```

```python
def sentiment_classifier(row):
    if row['overall'] > 3.0:
        return 'Positive'
    elif row['overall'] == 3.0:
        return 'Neutral'
    else:
        return 'Negative'
    
user_data['sentiment'] = user_data.apply(sentiment_classifier, axis=1)
user_data.head()
```
