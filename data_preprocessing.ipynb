```python
import json
import pandas as pd

path = './data_amazon_review/fine_food_reviews.csv'

user_data = pd.read_csv(path)
user_data.head()
```

### Based on user ratings in the 'overall' column, label reviews with Positive, Neutral, and Negative.

```python
sampled_data = user_data.sample(n=100000)
# sampled_data.sort_index().head()
sampled_data.reset_index()
sampled_data.head()
```

```python
def sentiment_classifier(row):
    if row['overall'] > 3.0:
        return 'Positive'
    elif row['overall'] == 3.0:
        return 'Neutral'
    else:
        return 'Negative'
    
user_data['sentiment'] = user_data.apply(sentiment_classifier, axis=1)
user_data.head()
```
